### Converting text data to numerical values

```python
calories_data.replace({'Gender':{'female':1,'male':0}}, inplace=True)
```

```python
calories_data
```

# **004 Separating features and Target**

```python
X = calories_data.drop(columns=['User_ID','Calories'], axis=1)
Y = calories_data['Calories']
```

```python
print(X)
```

```python
print(Y)
```

# **005 Splitting data into training and test data**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(Y.shape, Y_train.shape, Y_test.shape)
```

## **006 Training XGBoost Model**

```python
# Loading the model
model = XGBRegressor()
```

```python
# training the model with X_train, Y_train
model.fit(X_train,Y_train)
```

# ***007 Model Evaluation***

### Prediction on training data

```python
train_data_prediction = model.predict(X_train)
```

```python
train_data_prediction
```

```python
# Mean Absolute Error
MAE = metrics.mean_absolute_error(Y_train,train_data_prediction)
MAE
```

### Prediction on test data

```python
test_data_prediction = model.predict(X_test)
test_data_prediction
```

```python
# Mean Absolute Error
MAE = metrics.mean_absolute_error(Y_test,test_data_prediction)
MAE
```
